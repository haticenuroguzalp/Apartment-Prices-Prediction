```python
import missingno as msno
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df_may = pd.read_csv("apartments_pl_2024_05.csv")
df_june = pd.read_csv("apartments_pl_2024_06.csv")

df = pd.concat([df_may, df_june])
df.head()
```

## Aykırı Gözlem Analizi

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

# Alt ve üst eşikleri belirleyelim
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri filtreleyelim
df_outliers = df[(df["price"] < lower_bound) | (df["price"] > upper_bound)]
print(f"Aykırı fiyat değeri olan satır sayısı: {df_outliers.shape[0]}")
```

```python
df["log_price"] = np.log1p(df["price"])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["price"], bins=50, kde=True, ax=axes[0])
axes[0].set_title("Orijinal Fiyat Dağılımı")

sns.histplot(df["log_price"], bins=50, kde=True, ax=axes[1])
axes[1].set_title("Log Dönüşümlü Fiyat Dağılımı")

plt.show()
```

```python
lower_bound = np.percentile(df["squareMeters"], 1)  # En alt %1
upper_bound = np.percentile(df["squareMeters"], 99)  # En üst %99

# Baskılama uygula
df["squareMeters_capped"] = np.clip(df["squareMeters"], lower_bound, upper_bound)



fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x=df["squareMeters"], ax=axes[0])
axes[0].set_title("Orijinal Veriler")

sns.boxplot(x=df["squareMeters_capped"], ax=axes[1])
axes[1].set_title("Baskılanmış Veriler")

plt.show()
```

```python
Q1 = df["centreDistance"].quantile(0.25)
Q3 = df["centreDistance"].quantile(0.75)
IQR = Q3 - Q1

# Alt ve üst eşikleri belirleyelim
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri filtreleyelim
df_outliers = df[(df["centreDistance"] < lower_bound) | (df["centreDistance"] > upper_bound)]
print(f"Aykırı fiyat değeri olan satır sayısı: {df_outliers.shape[0]}")
```

```python
from scipy.stats.mstats import winsorize

# %1 alt ve %1 üst kesimi kırp
df["centreDistance_winsorized"] = winsorize(df["centreDistance"], limits=[0.01, 0.01])
```

```python
X = df[['latitude', 'longitude']].values

# Veriyi ölçekleyelim (DBSCAN ölçek duyarlıdır)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN Modeli (eps ve min_samples değerleri ayarlanabilir)
dbscan = DBSCAN(eps=0.3, min_samples=10)  # eps: komşuluk mesafesi, min_samples: minimum komşu sayısı
labels = dbscan.fit_predict(X_scaled)

# Aykırı noktaları belirleyelim (-1 olanlar outlier'dır)
df['dbscan_outlier'] = (labels == -1)

# Görselleştirme
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], c=df['dbscan_outlier'], cmap='coolwarm', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('DBSCAN ile Aykırı Konumlar')
plt.show()
```

```python
df['location_cluster'] = dbscan.labels_
```

## Eksik Gözlem Analizi

```python
msno.bar(df)
plt.show()

```

```python
missing_percent = df.isnull().sum() * 100 / len(df)
missing_df = pd.DataFrame({'Column': df.columns, 'Missing %': missing_percent})
missing_df = missing_df[missing_df["Missing %"] > 0].sort_values(by="Missing %", ascending=False)
print(missing_df)
```

```python
df.drop(columns=['condition','buildingMaterial'], inplace=True)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
df[['floor']] = imputer.fit_transform(df[['floor']])
```

```python
encoder = LabelEncoder()
df['type_encoded'] = encoder.fit_transform(df['type'])

imputer1 = KNNImputer(n_neighbors=5)
df['type_encoded'] = imputer1.fit_transform(df[['type_encoded']])
```

```python
imputer2 = KNNImputer(n_neighbors=5)  # 5 en yakın komşuyu kullan
df["buildYear"] = imputer2.fit_transform(df[["buildYear", "squareMeters", "floorCount"]])[:, 0]
```

```python
df['hasElevator'] = df['hasElevator'].fillna(df['hasElevator'].mode()[0])
```

```python
df.fillna({'collegeDistance': df['collegeDistance'].median()}, inplace=True)
```

```python
df['floorCount'] = df['floorCount'].fillna(df['floorCount'].mode()[0])
```

```python
df.fillna({'clinicDistance': df['clinicDistance'].median()}, inplace=True)
```

```python
df.fillna({'restaurantDistance': df['restaurantDistance'].median()}, inplace=True)
```

```python
df.fillna({'pharmacyDistance': df['pharmacyDistance'].median()}, inplace=True)
```

```python
df.fillna({'postOfficeDistance': df['postOfficeDistance'].median()}, inplace=True)
```

```python
df.fillna({'kindergartenDistance': df['kindergartenDistance'].median()}, inplace=True)
```

```python
df.fillna({'schoolDistance': df['schoolDistance'].median()}, inplace=True)
```

```python
df.drop(columns=['type','price','squareMeters','centreDistance','dbscan_outlier',], inplace=True)
```

```python
df['age'] = 2025 - df['buildYear']
```

## Standartizasyon İşlemleri

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = ['squareMeters_capped' , 'centreDistance_winsorized','longitude' , 'latitude' , 'location_cluster', 'schoolDistance', 'poiCount', 'clinicDistance', 'postOfficeDistance', 'kindergartenDistance' , 'restaurantDistance' , 'collegeDistance','pharmacyDistance' , 'age']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

## Label Encoding

```python
lbe = LabelEncoder()
for col in ["hasParkingSpace", "hasBalcony", "hasElevator", "hasSecurity", "hasStorageRoom"]:
    df[col] = lbe.fit_transform(df[col])
```

## One Hot Encoding

```python
ohe = OneHotEncoder(sparse_output=False, drop="first")  # İlk kategori drop edilir (dummy variable trap önleme)
categorical_cols = ['city', 'ownership']  
encoded = ohe.fit_transform(df[categorical_cols])

# Yeni dataframe oluştur
df_encoded = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(categorical_cols))
df_encoded.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)
df = pd.concat([df, df_encoded], axis=1).drop(columns=categorical_cols)
```

```python
df.drop(columns=['buildYear','id'], inplace=True)
```
